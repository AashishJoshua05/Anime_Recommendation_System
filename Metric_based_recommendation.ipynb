```python
takeClosest = lambda num,collection:min(collection,key=lambda x:abs(x-num))
for i in range(1,73517):
    x = dsr[(dsr['user_id'] == i)& (dsr[(dsr['user_id'] == i)]['rating']!=-1)]['anime_id'].to_numpy()
    count = 0
    for rati in x:
        for y in dsa[dsa['anime_id'] == rati]['episodes']:
            count = count+ int(y)
    y = count/len(x)
    no_epi = takeClosest(y,dsa[dsa['episodes']!='Unknown']['episodes'].to_numpy().astype(int))
    test = dsa[dsa['episodes'] == str(no_epi)].sort_values(['rating'], ascending = [False])
    itr = 0
    for names in test['name'].to_numpy():
        itr = itr+1
        print(f'{names}; ',end = '')
        if itr >3:
            break
    break
```
